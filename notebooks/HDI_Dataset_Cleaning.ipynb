# HDI Dataset Cleaning

### Data Sources
- HDI Dataset can be found [here](https://hdr.undp.org/data-center/documentation-and-downloads), titled [*All composite indices and components time series (1990-2022)*](https://hdr.undp.org/sites/default/files/2023-24_HDR/HDR23-24_Composite_indices_complete_time_series.csv).

### Importing Libraries:
`country_converter`: https://pypi.org/project/country-converter/

```python
import pandas as pd
import numpy as np
import country_converter as coco
```

### 1. Load Dataset:

```python
# Read in dataset, latin-1 encoding for custom characters
path_import = r'raw_data/HDR23-24_Composite_indices_complete_time_series.csv'
df_csv = pd.read_csv(path_import, encoding='latin-1')
```

```python
# Create copy of df to work with
df = df_csv.copy()

df.head(10)
```

```python
df.columns
```

```python
# Check all column headers
pd.set_option('display.max_columns', None)

df.head()
```

### 2. Filter Necessary Columns:

```python
# Keep country and relevant hdi columns
df_filtered = df[['country', 'hdi_2020', 'hdi_2021', 'hdi_2022']]
df_filtered
```

```python
# Check dtypes
df_filtered.dtypes
```

### 3. Check for Missing Values:

```python
# Check for NaN values
df_filtered[df_filtered.isna().any(axis=1)]
```

```python
# Drop NaN values
df_filtered = df_filtered.dropna()
```

### 4. Standardize Country Names
- Standardize country names for consistency, and remove non-country entries using `country_converter` package.

```python
# Extract country names
country_names = df_filtered['country']

# Standardize country names with country_converter
standard_names = coco.convert(country_names, to='name_short')

# Create dict mapping old and new country names
country_dict = dict(zip(country_names, standard_names))
```

```python
# Replace country with standardized names
df_standardized = df_filtered.copy()
df_standardized['country'] = df_filtered['country'].replace(country_dict)

# Drop 'not found' countries from df
df_standardized = df_standardized[df_standardized['country'] != 'not found']

# Check output
df_standardized
```

### 5. Rename Year Columns:

```python
# Rename year columns
df_renamed = df_standardized.rename(columns = {'hdi_2020': '2020', 
                                               'hdi_2021': '2021',
                                               'hdi_2022': '2022'})
df_renamed
```

### 6. Unpivot Years and HDI Values:

```python
# Unpivot columns and values to long format
df_melted = pd.melt(df_renamed, 
                    id_vars='country',
                    var_name='year',
                    value_name='hdi')

df_melted = df_melted.sort_values(['country', 'year']).reset_index(drop=True)

df_melted
```

### 7. Export to .csv:

```python
path_export = r'cleaned_data/hdi.csv'
df_melted.to_csv(path_export)
```
